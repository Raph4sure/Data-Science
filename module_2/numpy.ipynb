```python
#the face color was not set across all the subplots, we have to use a for loop to iterate and apply the color
fig, axs = plt.subplots(2,2, figsize=(3,3), facecolor='green', layout ='constrained')
fig.suptitle('Figure title')

#since axs is a numpy.ndarray, we have to iterate and make the changes to each plot
#remember ravel() fuunction in numpy?
#we use it to flatten the array so we can iterate through it

for ax in axs.ravel():
    ax.set_facecolor('yellow')

plt.show()
```

```python
#taking the above a step further

fig, axs = plt.subplots(2,2, figsize=(5,5), facecolor='green', layout='constrained')
fig.suptitle('Figure Title')

axes_title = ['Axes_1', 'Axes_2', 'Axes_3', 'Axes_4']
axes_color = ['yellow', 'purple', 'brown', 'orange']

#applying for each subplot

for idx, ax in enumerate(axs.ravel()):
    ax.set_facecolor(axes_color[idx])
    ax.set_title(axes_title[idx], loc ='center', fontsize='medium')
    ax.set_xlabel('x_label')
    ax.set_ylabel('y_label')

plt.show()
```

```python
#using more subplots

fig, axs = plt.subplots(1,3, figsize=(10,3), layout='constrained')
fig.suptitle('Comparing plots')

data = [data1, data1**2, data1**3] #list of datasets
labels = ['linear', 'quadratic', 'cubic'] #list of labels
colors = ['blue', 'brown', 'orange'] #list of line plot colors
axes_title = ['Linear', 'Quadratic', 'Cubic'] #list of subplot names

#1st line plot = x --> data1 (0th index), y--> data1 (0th index)
#2nd line plot = x --> data1 (0th index), y--> data1**2 (1st index)
#3rd line plot = x --> data1 (0th index), y--> data1**3 (2nd index)

for idx, ax in enumerate(axs.ravel()):
    ax.plot(data[0], data[idx],
            label=labels[idx],
            color=colors[idx])
    ax.axhline(2, ls='--', color='red') #ls = linestyle where -- means dashed lines
    ax.axvline(2, ls='--', color = 'black')

    ax.set_title(axes_title[idx], loc = 'center', fontsize=10)
    ax.set_xlabel('x_label')
    ax.set_ylabel('y_label')
    ax.legend()

plt.show()
```

```python
def my_plotter(axes, data1, data2, kind):
    if kind == 'line':
        out = axes.plot(data1, data2)
    else:
        out = axes.scatter(data1, data2)
    return out
```

```python
#using the helper function

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(5,3))

my_plotter(ax1, data1, data2, 'line')
my_plotter(ax2, data3, data4, 'scatter')

plt.show()
```

```python
def my_plot2(axes2, data1, data2, kind):
    if kind == "hist":
        chart = axes2.hist(data1, data2)
    elif kind == "scatter":
        chart = axes2.scatter(data1, data2)
    elif kind == "bar":
        chart = axes2.bar(data1, data2)
    else: chart = axes2.plot(data1, data2)
    
    return chart
```

```python
def discrete_uni_analysis(data):
    for col_name in data:
        print('\n', "*"*5, col_name, '*'*5)
        print(data[col_name].agg(['count', 'nunique', 'unique']))
        print("\n Value Counts: \n", data[col_name].value_counts())
        
def num_uni_analysis(data):
    for col_name in data:
        print('\n', "*"*5, col_name, '*'*5)
        print(data[col_name].agg(['count', 'min', 'max', 'mean', 'skew', 'median', 'std']))
```
